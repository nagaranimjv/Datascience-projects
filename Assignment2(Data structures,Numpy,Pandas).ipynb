```python
# 1.	Create a list of 5 integers. Perform and print the result of the following operations: 
##      append, extend, insert, remove, pop, clear, index, count, sort, and reverse.
```

```python
###### Creating list with 5 integers
list1 = [11,15,20,25,30]
print("Our created list: " , list1)
```

```python
### appending numbers to list1
list1.append(60)
list1
print("Our new list with appended value is: ", list1)
```

```python
# extend
list1.extend([70, 80])
print("Our list after extend:", list1)
```

```python
## insert (inserting 13 at position 2)
list1.insert(2, 13)
print("Our list after insert:", list1)
```

```python
# remove( removing number 40 from our list)
list1.remove(30)
print("our list after remove:", list1)
```

```python
# pop
popped_item = list1.pop()
print("our list after pop:", list1)
print("Popped item:", popped_item)
```

```python
# index
index_value = list1.index(15)
print("Index of 15 is:", index_value)
```

```python
# count( counts the how many times the given number is occured)
count_value = list1.count(20)
print("Count of 20:", count_value)
```

```python
# sort(gives the list in sorted order, default is ascending order)

list1.sort()
print("our liste after sort: ", list1)
```

```python
# reverse( it revese the list means the numbers in list will be in reverse order)
list1.reverse()
print("our list after reverse: ", list1)
```

```python
# clear( list will become empty)
list1.clear()
print("our list after clear: ", list1)
```

6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'


```python
### user input for marks
marks = int(input("enter the marks: "))
```

```python
### writing code for the conditions given
if marks >= 90:
   print("A")
else:
   if marks in range(75,89):
       print("B")
   else:
      if marks in range(50,74):
         print("C")
      else:   
        if marks <= 50:
           print("fail")
       
     
```
