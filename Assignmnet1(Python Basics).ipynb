Exercise 1: Prime Numbers

```python
# Write a Python program that checks whether a given number is prime or not. 
# A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
```

```python
num = int(input("Enter a number: "))

if num <= 1:
    print(num, "is not a prime number")
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```

Exercise 3: Squares of Even/Odd Numbers

Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

```python
#### writing program for the squares of even or odd numbers in range (10,20)
for num in range(10,20):
# check for the even number and finding the square of the number
    if num % 2 == 0:
       square_num = num**2
       print(f"this is even: {num}^2 = {square_num}")
# check for the odd number and calculate the square of the odd number
    else:
        if num % 2 != 0:
           square_num = num**2
           print(f"This is odd: {num}^2 = {square_num}")
        
```
