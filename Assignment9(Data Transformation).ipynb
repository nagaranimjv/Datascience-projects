●	Apply scaling techniques to numerical features:
a.	Standard Scaling   b. Min-Max Scaling


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Identify Numerical Columns
df_numerical = df.select_dtypes(include=['int64', 'float64']).columns
df_numerical
```

```python
# a) Standard Scaling
# Standard Scaling transforms data to have mean = 0 and std = 1.
# Creating standard scaler
standard_scaler = StandardScaler()

# Fit and transform
df_standard_scaled = df.copy()
df_standard_scaled[df_numerical] = standard_scaler.fit_transform(df[df_numerical])

print("Standard Scaled Data:")
print(df_standard_scaled.head())
```

```python
# b) Min-Max Scaling
# Min-Max Scaling transforms data to a range [0, 1].
# Create MinMacScaler
minmax_scaler = MinMaxScaler()

# Fit and transform MinMaxScaler
df_minmax_scaled = df.copy()
df_minmax_scaled[df_numerical] = minmax_scaler.fit_transform(df[df_numerical])

print("\nMin-Max Scaled Data:")
print(df_minmax_scaled.head())
```

●	Discuss the scenarios where each scaling technique is preferred and why


        Use Standard Scaling for Logistic Regression, SVM, PCA.
        Use Min-Max Scaling for Neural Networks, KNN.
        StandardScaler normalizes features to zero mean and unit variance, while MinMaxScaler rescales features into the range [0,1].
Scaling Techniques: When to Use Which?
        1. Standard Scaling (Z-Score Normalization)
           X(Scaled)  =  X-mu/std
           It transforms data to:
             Mean = 0
             Standard deviation = 1
   When data is approximately normally distributed:
        StandardScaler works best if features follow a bell-shaped distribution.
     For algorithms that assume centered data:   
        Algorithms like:
            Logistic Regression
            Linear Regression
            SVM
            PCA
            LDA
            benefit from zero-mean features for faster and better convergence. 
   When outliers exist but are not extreme:
           Standard scaling is less sensitive than Min-Max to moderate outliers.
   When comparing relative distances matters:
           Since it standardizes variance, it keeps meaningful relationships.
           
2. Min-Max Scaling (Normalization):
     X(scaled) = X-X(min)/x(max)-X(min)
     Transforms data into range: [0, 1
     
   when to use:
      When data does not follow Gaussian distribution,
   
      For algorithms based on distance calculations(KNN, Neural networks, K-means, Gradient Descent based models),
   
      When you need bounded inputs(Neural networks expects values between 0 and 1,
   
      When preserving relative magnitude is important( min-max keeps the original shape of the data.)
   
      When features have known min/max limits

   
  Why we use it:
  
       Ensures uniform scale.
       
       Speeds up learning in neural nets.
       
       Prevents dominance of large-scale features. 
      
Final conclusion: 

      Use Standard Scaling when data is normally distributed and used in linear models, and Min-Max Scaling when algorithms depend on distances or            require bounded input values.



#2. Encoding Techniques:
      

```python
# ●	Apply One-Hot Encoding to categorical variables with less than 5 categories.
## Identify categorical columns
df_category = df.select_dtypes(include=['object', 'category']).columns
    
##Select categorical columns with less than 5 unique values
low_cardinality_cols = [col for col in df_category if df[col].nunique() < 5]
print("Categorical columns with < 5 categories:")
print(low_cardinality_cols)
```

```python
# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=low_cardinality_cols, drop_first=True)
print("\nData after One-Hot Encoding:")
print(df_encoded.head())
```

```python
# Use Label Encoding for categorical variables. Data Exploration and Preprocessing:
df.describe()
```

```python
# Apply Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

df_labelencoded = df.copy()

for col in df_category:
    df_labelencoded[col] = label_encoder.fit_transform(df_labelencoded[col])

print("\nEncoded Data:")
print(df_labelencoded.head())
```

```python
## suggest new 2 features
   # 1. Workload Level (Hours × Education)
        # Feature: Education_Work_Level
   # 2. Net Capital (capital_gain - capital_loss)
        # Feature: Net_Capital
```

```python
# 1.Feature: Education_Work_Level
df['Education_Work_Level'] = df['education_num'] * df['hours_per_week']
```

```python
# 2.Feature: Net_Capital
df['Net_Capital'] = df['capital_gain'] - df['capital_loss']
```

```python
# ●	Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.
# checking skewness for 'capital gain
capital_gain_skew = df['capital_gain'].skew()
print("Skewness before log transformation for capital_gain is:", capital_gain_skew)
```

```python
# Apply log transformation
df['capital_gain_log'] = np.log1p(df['capital_gain'])
```

```python
# Check skewness after log transformation
# here we are creating new feature capital_gain_log to not distrubing the original value.
# and checking for the skewness after the log transformation
print("Skewness after log transformation for capital_gain is:", df['capital_gain_log'].skew())
```
