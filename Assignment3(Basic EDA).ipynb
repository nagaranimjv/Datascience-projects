### Boxplots

```python
#	Create boxplots for numerical variables to identify outliers and the interquartile range.
#	Discuss any findings, such as extreme values or unusual distributions.
```

## Boxplot for Volume

# Volume

#The distribution of volume is concentrated between 3 and 6 units, indicating that most transactions involve small quantities. However, several extreme values appear above 10 and reach beyond 30. These outliers suggest occasional bulk purchases that are very different from typical orders and contribute to a right-skewed distribution.

```python
# Calculate IQR for Volume
Q1 = df_numerical['Volume'].quantile(0.25)
Q3 = df_numerical['Volume'].quantile(0.75)
IQR = Q3 - Q1
print(f"{'Volume'} -> Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")

plt.figure(figsize=(5,6))
plt.boxplot(df_numerical['Volume'])
plt.title(f"Boxplot of {'Volume'}")
plt.ylabel('Volume')
plt.show()
```

## Boxplot for Avg Price

# Avg Price

#Avg Price shows a wide spread with many extreme values on the upper side. While most products are moderately priced, a few transactions involve very expensive items (above 50,000). These unusually high prices create strong positive skewness and increase variability in the dataset.

```python
# Calculate IQR for Avg Price
Q1 = df_numerical['Avg Price'].quantile(0.25)
Q3 = df_numerical['Avg Price'].quantile(0.75)
IQR = Q3 - Q1

print(f"{'Avg Price'} -> Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
plt.figure(figsize=(5,6))
plt.boxplot(df_numerical['Avg Price'])
plt.title(f"Boxplot of {'Avg Price'}")
plt.ylabel('Avg Price')
plt.show()
```

## Boxplot for Total Sales Value

#Total Sales Value

#The boxplot reveals that the median lies close to the lower quartile, while many extreme values appear far above the upper whisker (above 120,000). This indicates that most sales generate relatively low revenue, but a small number of large transactions dominate total sales, producing an uneven and highly right-skewed distribution.

```python
# Calculate IQR for Avg Price
Q1 = df_numerical['Total Sales Value'].quantile(0.25)
Q3 = df_numerical['Total Sales Value'].quantile(0.75)
IQR = Q3 - Q1
print(f"{'Total Sales Value'} -> Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
plt.figure(figsize=(5,6))
plt.boxplot(df_numerical['Total Sales Value'])
plt.title(f"Boxplot of {'Total Sales Value'}")
plt.ylabel('Total Sales Value')
plt.show()
```
