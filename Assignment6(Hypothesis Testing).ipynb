ESTIMATION AND CONFIDENCE INTERVALS

Background

In quality control processes, especially when dealing with high-value items, destructive sampling is a necessary but costly method to ensure product quality. The test to determine whether an item meets the quality standards destroys the item, leading to the requirement of small sample sizes due to cost constraints.


Scenario:

A manufacturer of print-heads for personal computers is interested in estimating the mean durability of their print-heads in terms of the number of characters printed before failure. To assess this, the manufacturer conducts a study on a small sample of print-heads due to the destructive nature of the testing process.


Data:

A total of 15 print-heads were randomly selected and tested until failure. The durability of each print-head (in millions of characters) was recorded as follows:
1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29


Assignment Tasks

a. Build 99% Confidence Interval Using Sample Standard Deviation

Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution for this task.


```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# The population standard deviation is unknown
# sample size is small <30
# we can go for t_test
```

```python
# given data
confidence = 0.99    # confidence interval
alpha =0.01         
n=15
degree_of_freedom = 14 #n-1
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
### sample statistics
mean = np.mean(data)

standard_deviation = np.std(data,ddof=1)
```

```python
print(mean)
print(standard_deviation)
```

```python
# t critical value
t_critical = stats.t.ppf(1-alpha/2,df=n-1)
t_critical
```

```python
# Standard error
standard_error = standard_deviation / np.sqrt(n)
standard_error
```

```python
# Margin of error
margin_error = t_critical * standard_error
margin_error
```

```python
# Confidence interval
ci_lower = mean - margin_error
ci_upper = mean + margin_error
```

```python
print(ci_lower)
print(ci_upper)
```

```python
print(f"Sample Size (n): {n}")
print(f"Sample Mean: {mean:.4f}")
print(f"Sample Std Dev: {standard_deviation:.4f}")
print(f"t Critical Value: {t_critical:.4f}")
print(f"99% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")
```

Practical Interpretation

In both scenarios, we are highly confident (99%) that the true mean number of characters printed before failure lies around 1.1 to 1.38 million characters.

Knowing the population standard deviation allows for a more precise estimate of the mean.

When the population standard deviation is unknown (which is most real-world cases), the t-distribution provides a safer, more conservative interval.

```python
# from the both scenarios, known and unknown values of standard deviation, we can define the null hypothesis scenarion.
# H0 = true mean durability of print-heads(mu_0) is equal to 1.2
# Ha = true mean durability of print-heads(mu_0) is not equal to 1.2
```

```python
# from the t stats we can conclude that the null hypothesis is whether fail to reject or reject
mu_0 = 1.2   # mean
sigma = 0.2  # standard deviation
# Sample statistics
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)
alpha = 0.01
```

```python
print("----- Sample Statistics -----")
print(f"Sample Size: {n}")
print(f"Sample Mean: {mean:.4f}")
print(f"Sample Std Dev: {std:.4f}")
```

```python
# t-test (sigma unknown)
t_stat, t_pvalue = stats.ttest_1samp(data, mu_0)
```

```python
t_stat,t_pvalue
```

```python
print("\n----- One-Sample t-test (σ unknown) -----")
print(f"t Statistic: {t_stat:.4f}")
print(f"p-value: {t_pvalue:.6f}")
```

```python
if t_pvalue < alpha:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
```

```python
# z-test (sigma known)
```

```python
se_z = sigma / np.sqrt(n)
z_stat = (mean - mu_0) / se_z
z_pvalue = 2 * (1 - stats.norm.cdf(abs(z_stat)))
```

```python
print("\n----- One-Sample z-test (σ known) -----")
print(f"z Statistic: {z_stat:.4f}")
print(f"p-value: {z_pvalue:.6f}")
```

```python
if z_pvalue < alpha:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
```
