```python
def outlier_capping(df_changed,columns):
    Q1= df_changed[columns].quantile(0.25)
    Q3= df_changed[columns].quantile(0.75)
    IQR= Q3-Q1
    lower_extreme= Q1-1.5*IQR
    upper_extreme= Q3+1.5*IQR
    df_changed[columns]=df_changed[columns].apply(lambda x: lower_extreme if x<lower_extreme else upper_extreme if x>upper_extreme else x)
for col in df_changed.select_dtypes(include=['int','float']).columns:
    outlier_capping(df_changed,col)
```
